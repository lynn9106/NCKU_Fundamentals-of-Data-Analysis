#### 處理缺失值
將test的資料合併一起處理

*   'Embarked'中只有train中兩筆資料缺失，選擇丟棄這兩筆資料
*   'Fare'中只有test中一筆資料缺失， 選擇填補整體票價的中位數
*   'Age' 中缺失值較多，且影響生存較大，選擇由乘客相同 name title 平均年齡填埔缺失年齡



````python
df_test = pd.read_csv('test.csv')
df_test.info()
# 合併train及test的資料 
df_data = df.append(df_test)

df_data['cabin_multiple'] = df.Cabin.apply(lambda x: 0 if pd.isna(x) else len(x.split(' ')))
df_data['cabin_adv'] = df.Cabin.apply(lambda x: str(x)[0])
df_data['numeric_ticket'] = df.Ticket.apply(lambda x: 1 if x.isnumeric() else 0)
df_data['name_title'] = df.Name.apply(lambda x: x.split(',')[1].split('.')[0].strip())

df_data.info()
````

````python
df_data.dropna(subset=['Embarked'],inplace = True)  #丟棄Embarked的兩筆資料
df_data['Fare'].fillna( df_data.Fare.median(), inplace=True ) #填補票價的中位數

NanName_title = df_data[df_data['Age'].isnull()].Name.apply(lambda x: x.split(',')[1].split('.')[0].strip())
NanName_title.value_counts()
````

確認Age缺失的乘客包含哪些name title 後，  
算出 ('Ms.''Miss.''Mrs.') 、 ('Mr.''Sir.''Major') 、 'Master' 、 'Dr.' 分別的平均年齡並填入缺失值

````python
#fill age  
mask = (df_data['Age'].isnull())&((df_data['Name'].str.contains('Ms.'))|(df_data['Name'].str.contains('Miss.'))|(df_data['Name'].str.contains('Mrs.')))
mask2 = ((df_data['Name'].str.contains('Ms.'))|(df_data['Name'].str.contains('Miss.'))|(df_data['Name'].str.contains('Mrs.')))
df_data.loc[mask,'Age'] = df_data.loc[mask,'Age'].fillna(df_data.loc[mask2,'Age'].median())

mask = (df_data['Age'].isnull())&((df_data['Name'].str.contains('Mr.'))|(df_data['Name'].str.contains('Sir.'))|(df_data['Name'].str.contains('Major')))
mask2 = ((df_data['Name'].str.contains('Mr.'))|(df_data['Name'].str.contains('Sir.'))|(df_data['Name'].str.contains('Major')))
df_data.loc[mask,'Age'] = df_data.loc[mask,'Age'].fillna(df_data.loc[mask2,'Age'].median())

mask = (df_data['Age'].isnull())&(df_data['Name'].str.contains('Master.'))
df_data.loc[mask,'Age'] = df_data.loc[mask,'Age'].fillna(df_data[df_data['Name'].str.contains('Master.')]['Age'].median())

mask = (df_data['Age'].isnull())&(df_data['Name'].str.contains('Dr\.'))
df_data.loc[mask,'Age'] = df_data.loc[mask,'Age'].fillna(df_data[df_data['Name'].str.contains('Dr\.')]['Age'].median())
````

#### 轉換數值資料(標準化)

使用 np.log() 函數進行對數轉換，加1來避免對於值為0的情況出現無限大的情況，使其更加接近正態分佈或者減少極端值


````python
# log norm of fare
df_data['norm_fare'] = np.log(df_data.Fare+1)
df_data.info()
````

#### 選取分析的資料欄位 與 轉換類別資料
* 轉換類別資料:   
使用LabelEncoder轉換  
(LabelEncoder 優點: 無需新增額外欄位、保留順序關係)

* 將 train 與 test 重新區分開來，並將以上處理過且認為影響生存的特徵放入train_x中  ，準備進行模型訓練

````python
from sklearn.preprocessing import LabelEncoder # 匯入 Label Encoder
le = LabelEncoder()                        # 創造 Label Encoder
for col in df_data[['Sex','Embarked','cabin_adv','name_title']]:
  le.fit(df_data[col])                         # 給予每個類別一個數值
  df_data[col] = le.transform(df_data[col])  # 轉換所有類別成為數值

Train = df_data[ pd.notnull(df_data.Survived) ]
Test = df_data[ pd.isnull(df_data.Survived) ]

train_x = Train[['Pclass','Sex','Age','SibSp','Parch','norm_fare','Embarked','cabin_adv','cabin_multiple','numeric_ticket','name_title']]                   # 取出訓練資料需要分析的資料欄位
train_y = Train['Survived']                       # 取出訓練資料的答案
````

#### 改善overfitting: 
**調整參數 max_depth**  
>因為較深的樹對訓練集的預測能力可能很好，但在新數據上可能較差，且較深的樹可能產生較多的分裂節點和葉節點，過於複雜的模型容易受到噪聲和無關特徵的影響，所以選擇通過限制樹的深度，提高模型的泛化能力。  

>數值為透過數次交叉驗證選擇的最佳數值

````python
# 模型訓練

from sklearn.model_selection import KFold             # 匯入 K 次交叉驗證工具
from sklearn.tree import DecisionTreeClassifier       # 匯入決策樹模型
from sklearn.metrics import accuracy_score            # 匯入準確度計算工具

kf = KFold(n_splits=5,                                # 設定 K 值
           random_state=1012,
           shuffle=True)
kf.get_n_splits(train_x)                              # 給予資料範圍

train_acc_list = []                                   # 儲存每次訓練模型的準確度
valid_acc_list = []                                   # 儲存每次驗證模型的準確度

for train_index, valid_index in kf.split(train_x):    # 每個迴圈都會產生不同部份的資料
    train_x_split = train_x.iloc[train_index]         # 產生訓練資料
    train_y_split = train_y.iloc[train_index]         # 產生訓練資料標籤
    valid_x_split = train_x.iloc[valid_index]         # 產生驗證資料
    valid_y_split = train_y.iloc[valid_index]         # 產生驗證資料標籤
    
    model_imp = DecisionTreeClassifier(max_depth=5 , random_state=1012) # 創造決策樹模型
    model_imp.fit(train_x_split, train_y_split)           # 訓練決策樹模型
    
    train_pred_y = model_imp.predict(train_x_split)       # 確認模型是否訓練成功
    train_acc = accuracy_score(train_y_split,         # 計算訓練資料準確度
                               train_pred_y)
    valid_pred_y = model_imp.predict(valid_x_split)       # 驗證模型是否訓練成功
    valid_acc = accuracy_score(valid_y_split,         # 計算驗證資料準確度
                               valid_pred_y)
    
    train_acc_list.append(train_acc)
    valid_acc_list.append(valid_acc)

print((
    'average train accuracy: {}\n' +
    '    min train accuracy: {}\n' +
    '    max train accuracy: {}\n' +
    'average valid accuracy: {}\n' +
    '    min valid accuracy: {}\n' +
    '    max valid accuracy: {}').format(
    np.mean(train_acc_list),                          # 輸出平均訓練準確度
    np.min(train_acc_list),                           # 輸出最低訓練準確度
    np.max(train_acc_list),                           # 輸出最高訓練準確度
    np.mean(valid_acc_list),                          # 輸出平均驗證準確度
    np.min(valid_acc_list),                           # 輸出最低驗證準確度
    np.max(valid_acc_list)                            # 輸出最高驗證準確度
)) 
````

#### Underfitting
透過調整參數，獲得極簡單的模型，使其underfitting  
1.   max_depth: 將最大深度設為1，即樹只有一個節點
2.   max_features: 設為1，表示每個節點只考慮一個特徵來進行分裂  


````python
# 模型訓練

from sklearn.model_selection import KFold             # 匯入 K 次交叉驗證工具
from sklearn.tree import DecisionTreeClassifier       # 匯入決策樹模型
from sklearn.metrics import accuracy_score            # 匯入準確度計算工具

kf = KFold(n_splits=5,                                # 設定 K 值
           random_state=1012,
           shuffle=True)
kf.get_n_splits(train_x)                              # 給予資料範圍

train_acc_list = []                                   # 儲存每次訓練模型的準確度
valid_acc_list = []                                   # 儲存每次驗證模型的準確度

for train_index, valid_index in kf.split(train_x):    # 每個迴圈都會產生不同部份的資料
    train_x_split = train_x.iloc[train_index]         # 產生訓練資料
    train_y_split = train_y.iloc[train_index]         # 產生訓練資料標籤
    valid_x_split = train_x.iloc[valid_index]         # 產生驗證資料
    valid_y_split = train_y.iloc[valid_index]         # 產生驗證資料標籤
    
    model_u = DecisionTreeClassifier( max_depth=1,max_features=1,random_state=1012) # 創造決策樹模型
    model_u.fit(train_x_split, train_y_split)           # 訓練決策樹模型
    
    train_pred_y = model_u.predict(train_x_split)       # 確認模型是否訓練成功
    train_acc = accuracy_score(train_y_split,         # 計算訓練資料準確度
                               train_pred_y)
    valid_pred_y = model_u.predict(valid_x_split)       # 驗證模型是否訓練成功
    valid_acc = accuracy_score(valid_y_split,         # 計算驗證資料準確度
                               valid_pred_y)
    
    train_acc_list.append(train_acc)
    valid_acc_list.append(valid_acc)

print((
    'average train accuracy: {}\n' +
    '    min train accuracy: {}\n' +
    '    max train accuracy: {}\n' +
    'average valid accuracy: {}\n' +
    '    min valid accuracy: {}\n' +
    '    max valid accuracy: {}').format(
    np.mean(train_acc_list),                          # 輸出平均訓練準確度
    np.min(train_acc_list),                           # 輸出最低訓練準確度
    np.max(train_acc_list),                           # 輸出最高訓練準確度
    np.mean(valid_acc_list),                          # 輸出平均驗證準確度
    np.min(valid_acc_list),                           # 輸出最低驗證準確度
    np.max(valid_acc_list)                            # 輸出最高驗證準確度
))  
````

#### 使用其他模型 與 ensemble (Soft Voting)



1.   [Naive Bayes](https://scikit-learn.org/stable/modules/naive_bayes.html#gaussian-naive-bayes):  
>基於貝葉斯定義和特徵條件獨立假設的分類器方法，通過計算在給定特徵下的條件概率，並利用貝葉斯定理來計算後驗概率，從而進行分類。  
>**訓練階段**，從樣本數據中估計特徵的條件概率；**預測階段**，根據給定特徵的條件概率和先驗概率計算後驗概率，並選擇具有最大後驗概率的類別作為預測結果。  
>**步驟 :** 數據預處理 -> 參數估計(每個特徵在每個類別下條件概率) -> 先驗概率估計 -> 後驗概率計算 -> 分類預測
2.   [K Nearest Neighbor](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html):  
>根據最接近的 K 個鄰居的標籤或值來預測新的樣本。  
>**步驟 :** 數據預處理 -> 距離度量(新樣本和訓練數據中每個樣本之間的距離) -> 鄰居選擇(K個) -> 標籤或值決策
3.   [Support Vector Classifier](https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html#sklearn.svm.SVC):  
>在特徵空間中找到一個最佳的超平面，來最大化不同類別之間的間隔，從而實現對樣本的有效分類。  
>**步驟 :** 數據預處理 -> 特徵空間映射(使用核函數（Kernel Function）將數據映射到一個高維特徵空間) -> 支持向量的選擇(距離最佳超平面最近的樣本) -> 超平面選擇(使得不同類別之間的間隔最大化) -> 預測
4.   [Soft Voting Classifier](https://scikit-learn.org/stable/modules/ensemble.html#voting-classifier):  
>一種集成學習方法。結合了多個不同的分類器（或稱為基學習器），並通過組合它們的預測結果來做出最終的分類決策。  
>對於多個基學習器的預測結果進行加權平均，權重可以根據基學習器的預測準確度進行調整，也就是計算每個基學習器對於某個樣本屬於不同類別的預測概率，然後將這些預測概率加權平均來獲得最終的預測概率，並根據最大概率來進行最終的分類預測。Soft Voting可以避免Hard Voting中平票的情況。





````python
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier

gnb = GaussianNB()
cv = cross_val_score(gnb,train_x,train_y,cv=5)
print("Naive_bayes mean: ")
print(cv.mean())
print()

knn = KNeighborsClassifier()
cv = cross_val_score(knn,train_x,train_y,cv=5)
print("KNN mean: ")
print(cv.mean())
print()

svc = SVC(probability = True)
cv = cross_val_score(svc,train_x,train_y,cv=5)
print("SVC mean: ")
print(cv.mean())
print()

voting_clf = VotingClassifier(estimators = [('knn',knn),('gnb',gnb),('svc',svc),('DescionTree',model_imp)], voting = 'soft', weights=[1,1,1,2])
print("voting_clf_soft mean :") 
print(cross_val_score(voting_clf,train_x,train_y,cv=5).mean())
````
